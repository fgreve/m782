```python
%%time
from tqdm import tqdm
from python_calamine import CalamineWorkbook

errors = []
for year in tqdm(years):
    print(str(year))
    path = os.path.join(path_archive, str(year), form_dict[year])

    Path(os.path.join(rawdata_path), str(year)).mkdir(parents=True, exist_ok=False)
    
    forms = [f for f in os.listdir(path) if f.find(".xls")>0]
    for f in forms:
        try:
            workbook = CalamineWorkbook.from_path(os.path.join(path, f))
            form_list = workbook.get_sheet_by_name(workbook.sheet_names[0]).to_python()
            with open(os.path.join(rawdata_path, str(year), f + ".txt"), "w", encoding="utf-8") as output:
                for item in form_list:
                    output.write(str(item) + "\n")
        
        except:
            errors.append(os.path.join(path, f))

    months = [f for f in os.listdir(path) if f.find(".")<0]
    for month in months:
        print(str(month))
        path_month = os.path.join(path, str(month))

        Path(os.path.join(rawdata_path), str(year), str(month)).mkdir(parents=True, exist_ok=False)
        
        forms = [f for f in os.listdir(path_month) if f.find(".xls")>0]
        for f in forms:
            print(f)
            
            try:
                workbook = CalamineWorkbook.from_path(os.path.join(path_month, str(f)))
                form_list = workbook.get_sheet_by_name(workbook.sheet_names[0]).to_python()
                with open(os.path.join(rawdata_path, str(year), str(month), str(f) + ".txt"), "w", encoding="utf-8") as output:
                    for item in form_list:
                        output.write(str(item) + "\n")
            except:
                errors.append(os.path.join(path_month, str(f)))                

        folders = [f for f in os.listdir(path_month) if f.find(".")<0]
        if 'PDF' in folders: folders.remove('PDF')   
        if 'PDF' in folders: folders.remove('PDFs') 
        for folder in folders:
            print(folder)
            Path(os.path.join(rawdata_path), str(year), str(month), folder).mkdir(parents=True, exist_ok=False)
            path_folder = os.path.join(path, str(month), str(folder))

            forms = [f for f in os.listdir(path_folder) if folder.find(".xls")>0]
            for f in forms:
                print(f)
                try:
                    workbook = CalamineWorkbook.from_path(os.path.join(path_folder, str(f)))
                    form_list = workbook.get_sheet_by_name(workbook.sheet_names[0]).to_python()
                    with open(os.path.join(rawdata_path, str(year), str(month), str(folder), str(f) + ".txt"), "w", encoding="utf-8") as output:
                        for item in form_list:
                            output.write(str(item) + "\n")
                except:
                    errors.append(os.path.join(rawdata_path, str(year), str(month), str(folder), str(f) + ".txt"))         
```
